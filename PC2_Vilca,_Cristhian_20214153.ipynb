# PC2: Limpieza y formateo de datos


## Lectura de datos



```python
# read table into pandas DF
import pandas as pd

HDIlink='https://github.com/CRVCh-16/PC2/raw/refs/heads/main/HDI.xlsx'
HDI=pd.read_excel(HDIlink)
```

```python
#la tabla a limpiar
HDI
```

## Limpieza de datos

```python
# Para conocer los nombres de las columnas que se van a actualizar
HDI.columns
```

```python
# Se crea una copia que será la usada para hacer la evaluación
HDI_copy=HDI.copy()
```

```python
# Se eliminan las columnas innecesarias
HDI_copy.drop(columns=["Back","Unnamed: 3","Unnamed: 5","Unnamed: 7","Unnamed: 9","Unnamed: 11","Unnamed: 12","Unnamed: 13","Unnamed: 14"],inplace=True)
HDI_copy
```

```python
# Se crean listas con los nombres de columnas antiguos y los nuevos
headings=list(HDI_copy.columns)
newheadings=["Country","Human_Development_Index","Life_expectancy_at_birth","Expected_years_of_schooling","Mean_years_of_schooling","Gross_national_income_per_capita"]
```

```python
# Se actualizan los nombres de las columnas
HDI_copy.rename(columns=dict(zip(headings,newheadings)),inplace=True)
HDI_copy
```

```python
# Se borran las primeras filas innecesarias
HDI_copy.drop(index=[0,1,2,3,4,5,6],inplace=True)
```

```python
# Se actualizan los índices del dataframe
HDI_copy.reset_index(drop=True,inplace=True)
HDI_copy
```

```python
# Se eliminan las últimas filas innecesarias
HDI_copy.drop(index=list(range(194,269)),inplace=True)
```

```python
# Se eliminan las filas intermedias innecesarias
HDI_copy.drop(index=[66,116,161],inplace=True)
```

```python
HDI_copy
```

```python
# Se hace la últim actualización de los índices del data frame

HDI_copy.reset_index(drop=True,inplace=True)
```

```python
# El resultado de la limpieza de datos es:

HDI_copy
```

## Formateo de datos

```python
# Se observan los tipos de datos de cada columna

HDI_copy.info()
```

```python
# Se hace el cambio de formato de tipo object a float64

HDI_copy['Human_Development_Index']=pd.to_numeric(HDI_copy.Human_Development_Index)
HDI_copy['Life_expectancy_at_birth']=pd.to_numeric(HDI_copy.Life_expectancy_at_birth)
HDI_copy['Expected_years_of_schooling']=pd.to_numeric(HDI_copy.Expected_years_of_schooling)
HDI_copy['Mean_years_of_schooling']=pd.to_numeric(HDI_copy.Mean_years_of_schooling)
HDI_copy['Gross_national_income_per_capita']=pd.to_numeric(HDI_copy.Gross_national_income_per_capita)

# Se observan los cambios

HDI_copy.info()
```

```python
# Entonces, el dataframe luego de la limpieza y formateo de datos queda como sigue:

HDI_copy
```

```python
# Finalmente, queda agregar una columna con el nivel de desarrollo humano

HDI_copy['Category']=pd.cut(HDI_copy.Human_Development_Index,
                           bins=[0,0.5495,0.7,0.799,1],
                           labels=['LOW HUMAN DEVELOPMENT','MEDIUM HUMAN DEVELOPMENT','HIGH HUMAN DEVELOPMENT','VERY HIGH HUMAN DEVELOPMENT'],
                           ordered=True)
HDI_copy
```
